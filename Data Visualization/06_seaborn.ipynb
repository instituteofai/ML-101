# Seaborn

Seaborn builds on top of `matplotlib`.

matplotlib = basic plotting (bars, pies, lines, scatter plots)  
seaborn = statistical visualization (higher level API than matplotlib)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Datasets/titanic_train.csv')
df.head()
```

```python
# Scatterplot
sns.relplot(x='Age', y='Fare', data=df)
```

```python
# Add color
sns.relplot(x='Age', y='Fare', hue='Pclass', data=df)
```

```python
# Change size
sns.relplot(x='Age', y='Fare', hue='Pclass', size='Parch', data=df)
```

Other interesting plots...

```python
sns.catplot(x='Pclass', y='Age', data=df, kind='box', hue='Sex')
```

```python
sns.catplot(x='Pclass', y='Age', data=df, kind='violin', split=True, hue='Sex')
```

```python
sns.distplot(df.Age)
```

```python
sns.jointplot(x='Age', y='Fare', data=df, kind="kde")
```

```python
plt.figure(figsize=(14, 7))
sns.heatmap(df.corr(), square=True, annot=True)
```
