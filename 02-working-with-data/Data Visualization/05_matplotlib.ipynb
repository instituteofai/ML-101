```python
import pandas as pd
import matplotlib.pyplot as plt

# Make our plots look nice
plt.style.use('fivethirtyeight')

%matplotlib notebook
```

```python
df = pd.read_csv('../Datasets/titanic_train.csv')
df.head()
```

```python
ages_male = df[df['Sex'] == 'male']['Age']
fares_male = df[df['Sex'] == 'male']['Fare']

ages_female = df[df['Sex'] == 'female']['Age']
fares_female = df[df['Sex'] == 'female']['Fare']

plt.scatter(ages_male, fares_male, color='green', label='Male')
plt.scatter(ages_female, fares_female, color='black', label='Female')

plt.xlabel('Age')
plt.ylabel('Fare')

plt.legend()

plt.show()
```

Figure = container holding plots  
Axes = actual plots

A figure can have multiple plots.

We have been using a 'stateful' approach.  
Get figure and axes using `plt.gcf()` and `plt.gca()` methods. 

Subplots = object-oriented way of doing this.

```python
fig, ax = plt.subplots()

ax.scatter(ages_male, fares_male, color='green', label='Male')
ax.scatter(ages_female, fares_female, color='black', label='Female')

ax.set_xlabel('Age')
ax.set_ylabel('Fare')

ax.legend()

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].scatter(ages_male, fares_male, color='green', label='Male')
ax[1].scatter(ages_female, fares_female, color='black', label='Female')

ax[0].set_xlabel('Age')
ax[1].set_xlabel('Age')
ax[0].set_ylabel('Fare')
ax[1].set_ylabel('Fare')

ax[0].legend()
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Sharing axis
fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].scatter(ages_male, fares_male, color='green', label='Male')
ax[1].scatter(ages_female, fares_female, color='black', label='Female')

ax[1].set_xlabel('Age')
ax[0].set_ylabel('Fare')
ax[1].set_ylabel('Fare')

ax[0].legend()
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Creating multiple figures
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.scatter(ages_male, fares_male, color='green', label='Male')
ax2.scatter(ages_female, fares_female, color='black', label='Female')

ax1.set_xlabel('Age')
ax2.set_xlabel('Age')
ax1.set_ylabel('Fare')
ax2.set_ylabel('Fare')

ax1.legend()
ax2.legend()

plt.show()
```
