##### On Combined Attributes
6.  In the last assignment Question 8, You would have found new combined attributes ( the new attributes don't need to correlate well with the label SalePrice).
Create a custom transformer class to add these to your dataset.

```python
from sklearn.base import BaseEstimator, TransformerMixin

# Get column indexes of four columns
# newhousing_df.columns.get_loc('HalfBath')
BsmtFullBath_x, BsmtHalfBath_x, FullBath_x, HalfBath_x = 17, 18, 19, 20
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):

    def __init__(self, total_bathroom=True):
        self.total_bathroom = total_bathroom
    
    def fit(self, X, y=None):
        return self
    
    def transform(self ,X,y=None):
#       Calculate total bathrooms and return
        total_bathroom_per_house = X[:, BsmtFullBath_x] + X[:, BsmtHalfBath_x] + X[:, FullBath_x] + X[: ,HalfBath_x]
        if self.total_bathroom:                  
            return np.c_[X, total_bathroom_per_house]
        else:
            return np.c_[X, total_bathroom_per_house]
        
# To get total number of bathrooms in a house
# housing_df['Total Bathrooom'] = housing_df['BsmtFullBath']  +  housing_df['BsmtHalfBath'] + housing_df['FullBath'] + housing_df['HalfBath']
# housing_df.drop('Total Bathrooom',axis=1)
```

```python
#Instantiate the class
attr_adder = CombinedAttributesAdder(total_bathroom=False)
housing_extra_col = attr_adder.transform(newhousing_df.values)
housing_extra_col.shape
# total_bathroom_per_house has been added to dataset
```

#### Build a full pipeline which:       
a. Transforms the attributes of the dataset and also

```python
housing_ames = pd.read_csv('train.csv')
housing_ames
```

```python
housings_labels = housing_ames[['SalePrice']]
housings_feautures = housing_ames[['OverallQual' ,'GrLivArea' , 'GarageCars','GarageArea', 'TotalBsmtSF','GarageYrBlt', 'Street','LotShape']]
housings_feautures.info()
```

```python
X_train , X_test , y_train , y_test = train_test_split(housings_feautures, housings_labels,test_size = 0.2,random_state = 42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler ,OneHotEncoder ,LabelEncoder , OrdinalEncoder
from sklearn import set_config
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
set_config(display='diagram')
#Pipeline for numerical attributes
num_pipeline = Pipeline([
                    ('median_imputer',SimpleImputer(strategy='median')),
                    ('stand_scaler' ,StandardScaler())
])

#Pipeline for categorical attributes
cat_pipeline = Pipeline([
                        ('one_hot_enc',OneHotEncoder(handle_unknown='ignore'))
])


num_pipeline
cat_pipeline
```

```python
numeric_attr = ['OverallQual' ,'GrLivArea' , 'GarageCars','GarageArea', 'TotalBsmtSF','GarageYrBlt']
pipeline = ColumnTransformer([
            ('numeric_transformers',num_pipeline, numeric_attr),
            ('categorical_transformer',cat_pipeline,['Street','LotShape'])
    
])
pipeline
```

```python
pipeline.fit_transform(X_train)
model_pipeline = Pipeline([('preprocessing',pipeline ), ('model', LinearRegression())])
model_pipeline.fit(X_test,y_test)
prediction = model_pipeline.predict(X_test)
rmse = mean_squared_error(y_test,prediction,squared=False)
r2 =  r2_score(y_test,prediction)
print(f'RMSE=>{rmse}')
print(f'r2=>{r2}')
```

###### Train the model with RandomForestRegressor and print the RMSE and r2_score. 

```python
pipeline_forest = Pipeline([
                        ('preprocessing', pipeline),('model', RandomForestRegressor())
])

pipeline_forest.fit(X_test,y_test)
for_prediction = pipeline_forest.predict(X_test) 
rmse = mean_squared_error(y_test,for_prediction,squared=False)
r2 =  r2_score(y_test,for_prediction)
print(f'RMSE=>{rmse}')
print(f'r2=>{r2}')
```

###### Support Vector Machine
Refer link below for more info on Support Vector Machine
https://towardsdatascience.com/support-vector-machine-introduction-to-machine-learning-algorithms-934a444fca47
In the SVM algorithm, we are looking to maximize the margin between the data points and the hyperplane. The loss function that helps maximize the margin is hinge loss.


https://towardsdatascience.com/svm-and-kernel-svm-fed02bef1200
As an example, lets consider two classes, apples and lemons.
Other algorithms will learn the most evident, most representative characteristics of apples and lemons, like apples are green and rounded while lemons are yellow and have elliptic form.
In contrast, SVM will search for apples that are very similar to lemons, for example apples which are yellow and have elliptic form. This will be a support vector. The other support vector will be a lemon similar to an apple (green and rounded). So other algorithms learns the differences while SVM learns similarities.

```python
## Using linear hyperparameter of SVC
from sklearn.svm import SVR
from  sklearn.metrics import accuracy_score 
# svr = SVR()

svr = Pipeline([
                        ('preprocessing', pipeline),('model', SVR())
                ])
svr.fit(X_train,y_train)
svr_prediction = svr.predict(X_test)
rmse = mean_squared_error(y_test,svr_prediction,squared=False)
r2 =  r2_score(y_test,svr_prediction)
print(f'RMSE=>{rmse}')
print(f'r2=>{r2}')
# print(accuracy_score(y_test,svr_prediction))
## We can see accuracy of SVC model is too low(.017)

##  Will use 'rbf'  as the kernel to check if accuracy increases or not?
```

```python
## Radial Basis Function (RBF)
# svr_rbf = SVR(kernel='rbf',C = 10, gamma = 0.1)
svr_rbf = Pipeline([
                        ('preprocessing', pipeline),('model', SVR(kernel='rbf',C = 10, gamma = 0.1))
                ])
svr_rbf.fit(X_train,y_train)
svr_rbf_prediction = svr_rbf.predict(X_test)
rmse = mean_squared_error(y_test,svr_rbf_prediction,squared=False)
r2 =  r2_score(y_test,svr_rbf_prediction)
print(f'RMSE=>{rmse}')
print(f'r2=>{r2}')
# print(accuracy_score(y_test,svc_rbf_prediction))
## Thers is no difference in R2 score in both Kernel RBF or Linear  .
```
