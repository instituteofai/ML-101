```python
import pandas as pd
df_train = pd.read_csv("train_otto.csv")
df_train.head()
```

```python
features = df_train.drop(['id', 'target'], axis=1)
labels = df_train[['target']]
```

```python
# Label encode target

from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
enc_labels = enc.fit_transform(labels.values.ravel())
enc_labels
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

preprocessing_pipeline = make_column_transformer(
  (
    make_pipeline(StandardScaler()),
    features
  ),
  sparse_threshold=0
)
```

```python
from sklearn.pipeline import Pipeline
#from sklearn.naive_bayes import GaussianNB
# from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn.svm import SVC
# from sklearn.linear_model import LogisticRegression

# model_pipeline = Pipeline([
#   ('preprocessing', preprocessing_pipeline),
#   ('model', GaussianNB())
# ])

clf=RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, enc_labels, test_size=0.2, random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
clf.predict_proba(X_test)
```
