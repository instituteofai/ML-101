```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
labels=df[['SalePrice']]
features=df.drop(['SalePrice'], axis=1)
```

```python
features.head()
```

```python
categorical_attributes = features.select_dtypes(include=[object]).columns.values
numerical_attributes = features.select_dtypes(exclude=[object]).columns.values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

numerical_imputer = SimpleImputer(strategy='median')
categorical_imputer = SimpleImputer(strategy='most_frequent')
ohe = OneHotEncoder(handle_unknown='ignore')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer

from sklearn import set_config
set_config(display='diagram')

numeric_pipeline = Pipeline([
    ('impute_numbers', numerical_imputer)
])

categorical_pipeline = Pipeline([
    ('impute_categories', categorical_imputer),
    ('one_hot_encoding', ohe)
])

transformer = ColumnTransformer([
    ('numeric_pipeline', numeric_pipeline, numerical_attributes),
    ('categorical_pipeline', categorical_pipeline, categorical_attributes)
])

pipeline = Pipeline([
    ('preprocessing', transformer),
    ('model', RandomForestRegressor())
])

pipeline
```

```python
pipeline.fit(X_train, y_train.values.ravel())
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

predictions = pipeline.predict(X_test)

mean_squared_error(np.log(y_test), np.log(predictions), squared=False)
```

```python
pipeline.fit(features, labels)
```

```python
test_df = pd.read_csv('test.csv')
test_df.head()
```

```python
predictions = pipeline.predict(test_df)
predictions
```

```python
predictions_df = pd.DataFrame({
    'Id': test_df.Id,
    'SalePrice': predictions
})

predictions_df.head()
```
